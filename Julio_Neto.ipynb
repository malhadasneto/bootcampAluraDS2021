### IMPORTAÇÃO LIBS E FUNÇÃO GRÁFICO

```
#Preparação inicial
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

pd.options.display.float_format = "{:.2f}".format

def plot_line(df, text_title, text_ylabel, size=(10,7), bbox=(1,0.86)):

  ax = df.plot (figsize=size)
  ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
  plt.title(text_title)
  plt.ylabel(text_ylabel)
  ax.legend(loc='lower left', bbox_to_anchor=bbox)
  plt.grid(axis='y')
  plt.show()
```

## IMPORTAÇÃO, LIMPEZA E PREPARAÇÃO DOS DATASETS
Decidimos aqui limitar a pesquisa aos últimos 10 anos. O último mês será jul/2020 pois estamos utilizando o banco de dados com datas por 'atendimento' e não queremos que eventuais atrasos no envio de informações nos últimos 3 meses atrapalhem nossa análise.

```
# Primeiramente vamos levantar os dados de gastos nos hospitais do SUS, divididos por estado

url = 'https://raw.githubusercontent.com/alura-cursos/agendamento-hospitalar/main/dados/A160324189_28_143_208.csv'
dados = pd.read_csv(url, encoding="ISO-8859-1", na_values='-',
                    skiprows = 3, sep=";", skipfooter=12, thousands=".", decimal=",", engine='python')
```

```
# Agora, vamos levantar os números de internações

url = 'https://raw.githubusercontent.com/malhadasneto/bootcampAluraDS2021/main/A110842189_28_143_208_internacoes-uf-ano-mes.csv'
dados_internacoes = pd.read_csv(url, encoding="ISO-8859-1", na_values='-', skiprows = 3, sep=";", skipfooter=12,
                    engine='python')
```

```
#Limitar o dataframe de Custos de Hospital a valores dos últimos 10 anos, até Junho/2020
custos_hospital_2010_2020 = dados.iloc[:, np.r_[0, 151:277]]
#ajustar indice
custos_hospital_2010_2020 = custos_hospital_2010_2020.set_index("Unidade da Federação")
#verificar valores vazios ou 0
custos_hospital_2010_2020.isnull().sum().sum()
#remover "números" do nome dos estados
for x in custos_hospital_2010_2020.index:
  custos_hospital_2010_2020 = custos_hospital_2010_2020.rename(index={x: x[3:]})
```

```
#Limitar o dataframe de Internações a valores dos últimos 10 anos, até Junho/2020
internacoes_2010_2020 = dados_internacoes.iloc[:,np.r_[0,37:163]]
#ajustar indice
internacoes_2010_2020 = internacoes_2010_2020.set_index("Unidade da Federação")
#verificar valores vazios ou 0
internacoes_2010_2020.isnull().sum().sum()
#remover "números" do nome dos estados
for x in internacoes_2010_2020.index:
  internacoes_2010_2020 = internacoes_2010_2020.rename(index={x: x[3:]})
```

```
#Valor gasto por pessoa internada
custo_por_pessoa_2010_2020 = custos_hospital_2010_2020
custo_por_pessoa_2010_2020 = custo_por_pessoa_2010_2020/internacoes_2010_2020
```

#Custo médio por internação: PR x Brasil
Vamos iniciar levantando o dado principal dessa análise: quanto custa em média a internação de uma pessoa.

```
custo_por_pessoa_2010_2020['Total'] = custo_por_pessoa_2010_2020.sum(axis=1)
custo_por_pessoa_2010_2020 = custo_por_pessoa_2010_2020.sort_values("Total", ascending=False)
custo_por_pessoa_2010_2020 = custo_por_pessoa_2010_2020.drop("Total", axis=1)

media = custo_por_pessoa_2010_2020.T.mean()
ax = media.plot (kind='bar', figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
plt.title("Custo médio por paciente internado - SUS - 2010 a 2020")
plt.grid(axis='y')
plt.xlabel("Fonte: DataSus/Tabnet Out/2020")
plt.ylabel("Em reais (R$)")
plt.ylim(0,1600)
plt.xticks(rotation=45, ha='right')
plt.show()
```

##Análise:

No gráfico a seguir mostramos a relação leitos Sus/10.000 hab. **O Paraná possui 16% mais leitos** que o segundo melhor estado do BR, Rio Grande do Sul.

Detalhe para um futuro trabalho: Santa Catarina apresenta o 2o. maior custo médio por paciente internado, porém está **apenas em 8. lugar** em leitos Sus/10.000 hab.

```
#proporção gastos/leitos (quanto se gasta por leito)
#apenas um exercício para ver se o PR continua na 'liderança'
dados_utis['Leitos de UTI']
custos_hosp_2019 = custos_hospital_2010_2020.T[108:120].T
custos_hosp_2019['Total'] = custos_hosp_2019.sum(axis=1)
#quanto custa cada leito?
dados_utis["Gastos / Leitos SUS"] = custos_hosp_2019['Total'] / dados_utis["Leitos de UTI SUS"]
```

```
#por MUNICIPIO DO PR - vamos ver se achamos um município 'culpado' ou se é geral
url_internacoes = "https://raw.githubusercontent.com/malhadasneto/bootcampAluraDS2021/main/internacoes%20atendimento%20segundo%20municipio%20PR.csv"
url_gastos = "https://raw.githubusercontent.com/malhadasneto/bootcampAluraDS2021/main/procedimentoshospitalares%20por%20atendimento%20por%20municipio%20PR.csv"

dados_internacoes_por_mun_pr = pd.read_csv(url_internacoes, encoding="UTF-8", na_values='-',
                    sep=";", thousands=".", decimal=",", engine='python')
dados_gastos_hosp_por_mun_pr = pd.read_csv(url_gastos, encoding="UTF-8", na_values='-',
                    sep=";", thousands=".", decimal=",", engine='python')

#Limitar o dataframe de Custos de Hospital e Internações a valores dos últimos 10 anos, até Junho/2020
dados_gastos_hosp_por_mun_pr = dados_gastos_hosp_por_mun_pr.iloc[:, np.r_[0, 25:151]]
dados_gastos_hosp_por_mun_pr = dados_gastos_hosp_por_mun_pr.set_index("Município")
for x in dados_gastos_hosp_por_mun_pr.index:
  dados_gastos_hosp_por_mun_pr = dados_gastos_hosp_por_mun_pr.rename(index={x: x[7:]})

dados_internacoes_por_mun_pr = dados_internacoes_por_mun_pr.iloc[:, np.r_[0, 9:135]]
dados_internacoes_por_mun_pr = dados_internacoes_por_mun_pr.set_index("Município")
for x in dados_internacoes_por_mun_pr.index:
  dados_internacoes_por_mun_pr = dados_internacoes_por_mun_pr.rename(index={x: x[7:]})

  #criar dataframe gastos / pacientes
custo_por_pessoa_mun_pr = dados_gastos_hosp_por_mun_pr.copy()
custo_por_pessoa_mun_pr = dados_gastos_hosp_por_mun_pr / dados_internacoes_por_mun_pr
```

```
#Vamos ver quais municípios tem mais impacto no total do Paraná
dados_gastos_hosp_por_mun_pr['Total'] = dados_gastos_hosp_por_mun_pr.sum(axis=1)
dados_gastos_hosp_por_mun_pr = dados_gastos_hosp_por_mun_pr.sort_values("Total", ascending=False)
```
